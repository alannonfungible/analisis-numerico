Un programa que resuelva un sistema de dos bandas, puede ser el siguiente:

```python
import numpy as np

def resolver_sistema_dos_bandas(A, b):
    """
    Resuelve un sistema de ecuaciones lineales con una matriz de dos bandas usando eliminación de Gauss.

    Parámetros:
    A : Matriz de coeficientes.
    b : Vector de términos independientes.

    Retorna:
    x : Solución del sistema de ecuaciones.
    """
    n = len(b)
    # Eliminación hacia adelante
    for i in range(n):
        # Normalizar la fila i
        factor = A[i, i]
        A[i, :] /= factor
        b[i] /= factor

        # Eliminar la variable i de las filas siguientes
        for j in range(i + 1, n):
            factor = A[j, i]
            A[j, :] -= factor * A[i, :]
            b[j] -= factor * b[i]

    # Sustitución hacia atrás
    x = np.zeros(n)
    for i in range(n - 1, -1, -1):
        x[i] = b[i] - np.dot(A[i, i + 1:], x[i + 1:])

    return x

# Ejemplo de uso

A = np.array([
        [5, -1, 0, 0, 0],
        [1, 4, -1, 0, 0],
        [0, 1, 3, -1, 0],
        [0, 0, 1, 2, -1],
        [0, 0, 0, 1, 1]
    ], dtype=float)

# Definir el vector de términos independientes
b = np.array([0, 1, 2, 2, 3], dtype=float)

# Resolver el sistema
x = resolver_sistema_dos_bandas(A, b)
print("Solución del sistema:", x)
```
