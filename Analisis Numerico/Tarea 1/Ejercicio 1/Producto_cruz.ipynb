Un programa que aplique un producto cruz entre dos vectores tridimensionales puede ser el siguiente:

```python
def producto_cruz(vector1, vector2):
    """
    Calcula el producto cruz entre dos vectores tridimensionales.

    Parámetros:
    vector1 : Lista que representa el primer vector tridimensional [x1, y1, z1].
    vector2 : Lista que representa el segundo vector tridimensional [x2, y2, z2].

    Retorna:
    list: El producto cruz de los dos vectores, también como una lista [x3, y3, z3].
    """

    # El producto cruz de dos vectores (a, b, c) y (d, e, f) es un vector (x, y, z) donde:
    # x = b*f - c*e
    # y = c*d - a*f
    # z = a*e - b*d

    x = vector1[1] * vector2[2] - vector1[2] * vector2[1]
    y = vector1[2] * vector2[0] - vector1[0] * vector2[2]
    z = vector1[0] * vector2[1] - vector1[1] * vector2[0]

    # Devolvemos el vector resultado del producto cruz
    return [x, y, z]

#Ejemplo de uso
vector1 = [1, 0, 0]
vector2 =[0 ,1 ,2]

# Calculamos el producto cruz entre los dos vectores
resultado = producto_cruz(vector1, vector2)

# Mostramos el resultado del producto cruz
print(f"El producto cruz entre {vector1} y {vector2} es: {resultado}")
```
