La norma 2 se define de la siguiente manera:
$$\left\Vert A\right\Vert _{2}:=\sqrt{\sum_{j=1}^{n}a_{ij}^{2}}$$

La función para calcularla es:


```python
#Importamos numpy
import numpy as np
#Definimos una matriz de ejemplo
A=[[1, 5,8 ],[4, -5, 10],[9,15,-2]]


def inducidaDos(A):
  """

  Función para obtener la norma dos de una matriz.

  Parámetros
  -------------
  A : La matriz del cual sacar la norma.

  Regresa
  -------------
  norma : La norma dos de la matriz.
  """

  A_transpuesta = [list(x) for x in zip(*A)]
  producto = []
  for i in range(len(A_transpuesta)):
        columna = []
        for j in range(len(A[0])):
            sum = 0
            for k in range(len(A)):
                sum += A_transpuesta[i][k] * A[k][j]
            columna.append(sum)
        producto.append(columna)

  eigenvalores = np.linalg.eigvals(producto)
  espectral_norm = np.sqrt(max(abs(eigenvalores)))

  return espectral_norm



print("La norma dos de la matriz es:",inducidaDos(A))
```
