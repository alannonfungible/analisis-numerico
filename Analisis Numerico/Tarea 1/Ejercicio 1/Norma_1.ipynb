La norma uno esta definida como:


$$\left\Vert A\right\Vert _{1}:=\underset{1\leq j\leq n}{max}\sum_{i=1}^{n}\left|a_{ij}\right|$$

La función para calcularla es:

```python
# Definimos una matriz de ejemplo
A = [[3, -1, 4],[-5, 0, 2],[1, -2, 6]]

def inducidaUno(A):
  """

  Función para obtener la norma inducida por la norma uno de una matriz.

  Parámetros
  -------------
  A : matrix
          La matriz del cual sacar la norma.

  Regresa
  -------------
  norma : int
          La norma inducida por la norma uno de la matriz.
  """

  #  Inicializanos la lista de sumas a cero
  sumas = [0] * len(A[0])

  # Se suman las columnas de la matriz

  for fila in A:
      for i, valor in enumerate(fila):
        sumas[i] += abs(valor)

  return max(sumas)

print("La norma matricial inducida por la norma uno es:", inducidaUno(A))
```
