Un programa que realice la transpuesta de una matriz, puede ser el siguiente:

```python
def transponer_matriz(matriz):
    """
    Esta función toma una matriz (lista de listas) y devuelve su transpuesta.

    Parámetros:
    matriz : La matriz a transponer.

    Retorna:
    list : La matriz transpuesta.
    """
    # Verificar si la matriz está vacía
    if not matriz:
        return []

    # Obtener el número de filas y columnas de la matriz original
    filas = len(matriz)
    columnas = len(matriz[0])

    # Crear una nueva matriz vacía con dimensiones invertidas
    matriz_transpuesta = [[None] * filas for _ in range(columnas)]

    # Llenar la matriz transpuesta
    for i in range(filas):
        for j in range(columnas):
            matriz_transpuesta[j][i] = matriz[i][j]

    return matriz_transpuesta

# Ejemplo de uso
matriz_original = [
    [1, 2, 3],
    [4, 5, 6],
    [7, 8, 9]
]

print("Matriz original:")
for fila in matriz_original:
    print(fila)

matriz_transpuesta = transponer_matriz(matriz_original)

print("\nMatriz transpuesta:")
for fila in matriz_transpuesta:
    print(fila)
```
