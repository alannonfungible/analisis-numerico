Un programa que aplique una rotación de un vector tridimensional mediante una matriz de rotación, puede ser el siguiente:  

```python
import math

def matriz_rotacion_x(angulo):
    """
    Crea una matriz de rotación en torno al eje X.

    Parámetros:
    angulo : El ángulo de rotación en radianes.

    Retorna:
     list: La matriz de rotación 3x3.
    """
    return [
        [1, 0, 0],
        [0, math.cos(angulo), -math.sin(angulo)],
        [0, math.sin(angulo), math.cos(angulo)]
    ]

def matriz_rotacion_y(angulo):
    """
    Crea una matriz de rotación en torno al eje Y.

    Parámetros:
    angulo (float): El ángulo de rotación en radianes.

    Retorna:
    list : La matriz de rotación 3x3.
    """
    return [
        [math.cos(angulo), 0, math.sin(angulo)],
        [0, 1, 0],
        [-math.sin(angulo), 0, math.cos(angulo)]
    ]

def matriz_rotacion_z(angulo):
    """
    Crea una matriz de rotación en torno al eje Z.

    Parámetros:
    angulo (float): El ángulo de rotación en radianes.

    Retorna:
    list : La matriz de rotación 3x3.
    """
    return [
        [math.cos(angulo), -math.sin(angulo), 0],
        [math.sin(angulo), math.cos(angulo), 0],
        [0, 0, 1]
    ]

def rotar_vector(vector, matriz_rotacion):
    """
    Aplica una rotación a un vector tridimensional utilizando una matriz de rotación.

    Parámetros:
    vector : El vector tridimensional a rotar.
    matriz_rotacion : La matriz de rotación 3x3.

    Retorna:
    list: El vector rotado.
    """
    return [
        sum(vector[j] * matriz_rotacion[i][j] for j in range(3))
        for i in range(3)
    ]

# Ejemplo de uso
vector = [1, 0, 0]
angulo = math.radians(90)  # Convertir 90 grados a radianes

# Seleccionar la matriz de rotación deseada (en este caso, rotación en torno al eje Z)
matriz_rot = matriz_rotacion_z(angulo)

vector_rotado = rotar_vector(vector, matriz_rot)

print("Vector original:", vector)
print("Vector rotado:", vector_rotado)
```
