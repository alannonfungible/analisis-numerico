La norma infinitio esta definida como:


$$\left\Vert A\right\Vert _{\infty}:=\underset{1\leq i\leq n}{max}\sum_{j=1}^{n}\left|a_{ij}\right|$$

La función para calcularla es:

```python
#
import numpy as np
# Definimos una matriz de ejemplo
A = [[3, -1, 4],[-5, 0, 2],[1, -2, 6]]
def inducidaInfinito(A):
  """

  Función para obtener la norma inducida por la norma del infinito

  Parámetros
  -------------
  A : La matriz del cual sacar la norma.

  Regresa
  -------------
  norma : la norma inducida por la norma del infinito
  """
  sumas = []
  # Se suman los valores absolutos de los renglones de la matriz
  for i, fila in enumerate(A):
    suma_renglon = sum(abs(valor) for valor in fila)
    sumas.append(suma_renglon)
  return max(sumas)

print("La norma matricial inducida por la norma del infinito es:",inducidaInfinito(A))
```
