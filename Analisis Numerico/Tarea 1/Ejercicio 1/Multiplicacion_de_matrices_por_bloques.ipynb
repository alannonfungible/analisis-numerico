Un programa que realice la multiplicación de matrices por bloque, puede ser el siguiente:

```python
def multiplicar_bloques(A, B, tam_bloque):
    """
    Realiza la multiplicación de matrices por bloques utilizando listas.

    Parámetros:
    A : La primera matriz.
    B : La segunda matriz.
    tam_bloque: El tamaño de los bloques.

    Retorna:
     list: La matriz resultante de la multiplicación.
    """
    # Verificar que las dimensiones sean compatibles para la multiplicación
    assert len(A[0]) == len(B), "Las matrices no son compatibles para la multiplicación."

    # Dimensiones de las matrices
    n = len(A)
    m = len(B[0])
    p = len(A[0])

    # Crear la matriz resultante
    C = [[0] * m for _ in range(n)]

    # Multiplicación por bloques
    for i in range(0, n, tam_bloque):
        for j in range(0, m, tam_bloque):
            for k in range(0, p, tam_bloque):
                # Definir los límites de los bloques
                i_max = min(i + tam_bloque, n)
                j_max = min(j + tam_bloque, m)
                k_max = min(k + tam_bloque, p)

                # Multiplicar los bloques y acumular el resultado
                for ii in range(i, i_max):
                    for jj in range(j, j_max):
                        for kk in range(k, k_max):
                            C[ii][jj] += A[ii][kk] * B[kk][jj]

    return C

# Ejemplo de uso
A = [
    [1, 2, 3],
    [4, 5, 6],
    [7, 8, 9]
]

B = [
    [9, 8, 7],
    [6, 5, 4],
    [3, 2, 1]
]

tam_bloque = 2

print("Matriz A:")
for fila in A:
    print(fila)

print("\nMatriz B:")
for fila in B:
    print(fila)

C = multiplicar_bloques(A, B, tam_bloque)

print("\nMatriz resultante C:")
for fila in C:
    print(fila)
```
